## Configuration

```python
PROJECT_ROOT = Path("..")

RAW_DATA_DIR = PROJECT_ROOT / "data" / "raw"
PROCESSED_DATA_DIR = PROJECT_ROOT / "data" / "processed"

RAW_STOCK_DIR = RAW_DATA_DIR / "stocks"
RAW_INDEX_DIR = RAW_DATA_DIR / "indices"
RAW_ETF_DIR = RAW_DATA_DIR / "etfs"
RAW_COMMODITY_DIR = RAW_DATA_DIR / "commodities"
RAW_FOREX_DIR = RAW_DATA_DIR / "forex"
RAW_MACRO_DIR = RAW_DATA_DIR / "macro"
RAW_NEWS_DIR = RAW_DATA_DIR / "news"

PROCESSED_STOCK_DIR = PROCESSED_DATA_DIR / "stocks"
PROCESSED_INDEX_DIR = PROCESSED_DATA_DIR / "indices"
PROCESSED_ETF_DIR = PROCESSED_DATA_DIR / "etfs"
PROCESSED_COMMODITY_DIR = PROCESSED_DATA_DIR / "commodities"
PROCESSED_FOREX_DIR = PROCESSED_DATA_DIR / "forex"
PROCESSED_MACRO_DIR = PROCESSED_DATA_DIR / "macro"
PROCESSED_NEWS_DIR = PROCESSED_DATA_DIR / "news"

directories = [
    PROCESSED_STOCK_DIR,
    PROCESSED_INDEX_DIR,
    PROCESSED_ETF_DIR,
    PROCESSED_COMMODITY_DIR,
    PROCESSED_FOREX_DIR,
    PROCESSED_MACRO_DIR,
    PROCESSED_NEWS_DIR
]

for directory in directories:
    directory.mkdir(parents=True, exist_ok=True)

print("Directories Ready")
```

# Load Dataset

```python
def load_dataset(folder: Path):

    datasets = {}

    csv_files = sorted(folder.glob("*.csv"))

    print(f"Found {len(csv_files)} files\n")

    for file in csv_files:

        try:

            df = pd.read_csv(file)

            datasets[file.stem] = df

            print(f"{file.name:<25} {df.shape}")

        except Exception as e:

            print(file.name, e)

    return datasets
```

```python
stock_data = load_dataset(RAW_STOCK_DIR)

index_data = load_dataset(RAW_INDEX_DIR)

etf_data = load_dataset(RAW_ETF_DIR)

commodity_data = load_dataset(RAW_COMMODITY_DIR)

forex_data = load_dataset(RAW_FOREX_DIR)

macro_data = load_dataset(RAW_MACRO_DIR)

news_data = load_dataset(RAW_NEWS_DIR)
```

## Standardize Column Names

```python
def clean_columns(df):

    df.columns = (
        df.columns
        .str.strip()
        .str.replace(" ", "_")
        .str.replace("-", "_")
    )

    return df
```

```python
for datasets in [
    stock_data,
    index_data,
    etf_data,
    commodity_data,
    forex_data,
    macro_data,
    news_data
]:

    for key in datasets:

        datasets[key] = clean_columns(datasets[key])

print("Columns Standardized")
```
